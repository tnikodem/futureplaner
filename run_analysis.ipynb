```python
import yaml

from bokeh.plotting import show
from bokeh.io import output_notebook
output_notebook()

from fup.utils import get_module_start_values, run_toys
from fup.plot_utils import create_plot
```

```python
with open("config_template.yaml", "r") as f:
    config = yaml.load(f.read(), Loader=yaml.SafeLoader)
config["random"] = False
config["mean_inflation"] = 0
get_module_start_values(config=config)    
```

```python
with open("config_template.yaml", "r") as f:
    config = yaml.load(f.read(), Loader=yaml.SafeLoader)
df, df_stats = run_toys(config=config, runs=1000)
```

```python
# Your total assets (inflation corrected)
p = create_plot(df=df, column="assets_cor", name="Assets", unit="€", color="#4070f5")
show(p)
```

```python
# Yearly net expenses (without tax&insurance, inflation corrected)
p = create_plot(df=df, column="expenses_net_cor", name="Yearly expenses", unit="€", color="#990000",
               legend_location="bottom_right")
show(p)
```

```python
# Money Level
# +10: You should spend much more money
#   0: Your spendings are ok
# -10: You must only spend the absolute necessary
p = create_plot(df=df, column="money_level", name="Money Level", unit="", color="#007700",
               legend_location="bottom_right")
show(p)
```
